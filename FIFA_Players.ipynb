```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Preprocess the dataset
df = pd.read_csv(r"C:\Users\RAMEESHA\Downloads\fifa_data.csv")
# Convert 'Wage' and 'Height' columns to numeric values for analysis
def convert_wage(value):
    try:
        # Extract numeric value from wage and convert to integer
        return int(value[1:-1].replace('K', '')) * 1000 if "K" in value else int(value[1:-1].replace('M', '')) * 1000000
    except:
        return 0

def convert_height(value):
    if isinstance(value, str) and "'" in value:
        # Convert height from feet and inches to centimeters
        feet, inches = map(int, value.split("'"))
        return feet * 30.48 + inches * 2.54  # Convert feet and inches to centimeters
    return None

# Applying conversions
df['Wage'] = df['Wage'].apply(convert_wage)
df['Height'] = df['Height'].apply(convert_height)

# Drop rows with null values in relevant columns for clean analysis
cleaned_df = df.dropna(subset=['Nationality', 'Wage', 'Height', 'Club', 'Preferred Foot'])

# Drop irrelevant columns for analysis (e.g., image links, extra details)
columns_to_drop = ['Unnamed: 0', 'ID', 'Photo', 'Flag', 'Club Logo', 'Value', 'Release Clause', 'Loaned From', 
                   'Contract Valid Until', 'Joined', 'Real Face', 'Body Type', 'Work Rate']
cleaned_df = cleaned_df.drop(columns=columns_to_drop)

# Display the cleaned dataset information and first few rows
cleaned_df.info(), cleaned_df.head()
```

Insights
1. Country with the Most Number of Players

```python
country_counts = df['Nationality'].value_counts()
country_counts
```

```python
most_players_country = country_counts.idxmax()
print(f"Country with the most number of players: {most_players_country}")
```

```python
top_5_countries = country_counts.head(5)
```

```python
top_5_countries
```

```python
plt.figure(figsize=(10, 6))
top_5_countries.plot(kind='bar', color='blue')
plt.title('Top 5 Countries with the Most Number of Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()
```

Insights: The country with the most FIFA players is England (1,657 players), followed by Germany (1,195), Spain (1,071), Argentina (935), and France (911). The bar chart above visualizes the top 5 countries.

```python
highest_salary_player = df.loc[df['Wage'].idxmax()]
print(f"Player with the highest salary: {highest_salary_player['Name']}, Salary: {highest_salary_player['Wage']}")
```
